```python
import cv2
import numpy as np
import json
from urllib.request import urlopen

def print_longitude():
    url="https://ipinfo.io/json"
    response = urlopen(url)
    data = json.load(response)
    print(data["loc"])
    

prototxt_path = 'Model/MobileNetSSD_deploy.prototxt'
model_path = 'Model/MobileNetSSD_deploy.caffemodel'
min_confidence = 0.5

classes = ["person"]
colors = (255, 0, 0)
net = cv2.dnn.readNetFromCaffe(prototxt_path, model_path)
cap = cv2.VideoCapture(0)

counter = 0
while True:
    
    _, image = cap.read() 
    height, width = image.shape[0], image.shape[1]
    blob = cv2.dnn.blobFromImage(cv2.resize(image,(300,300)), 0.007, (300,300),130)

    net.setInput(blob)
    detected_objects = net.forward()
    
    
    confidence = detected_objects[0][0][0][2]

    if confidence > min_confidence:
        class_index = int(detected_objects[0,0,0,1])

        upper_left_x = int(detected_objects[0,0,0,3]*width)
        upper_left_y = int(detected_objects[0,0,0,4]*height)
        lower_right_x = int(detected_objects[0,0,0,5]*width)
        lower_right_y = int(detected_objects[0,0,0,6]*height)


        prediction_text = f"{classes[0]}:{confidence:.2f}%"
        cv2.rectangle(image,(upper_left_x, upper_left_y),(lower_right_x, lower_right_y), colors[0],3)
        newImage = cv2.putText(image, prediction_text, (upper_left_x, 
                    upper_left_y-15 if upper_left_y>30 else upper_left_y+15), cv2.FONT_HERSHEY_SIMPLEX, 0.5, 
                    colors[0],1)
        print(detected_objects[0][0][0][1])
        
        if detected_objects[0][0][0][1] == 15 and counter%15 == 0: 
            
            new_temp = "Faces/"+str(int(counter/15))+".jpg"
            
            cv2.imwrite(new_temp, newImage)
            print_longitude()
        counter += 1 
                
    cv2.imshow('Image',image)
    

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows()
cap.release()
```
